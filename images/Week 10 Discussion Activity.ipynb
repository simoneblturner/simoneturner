### PageRank

 To find PageRank in practice we combine two mechanisms:

1.  We give each node an amount of “status” (or PageRank). At each time step the node distributes that PageRank to other nodes.

2. At each time step, each node distributes a portion of PageRank depending on the damping parameter (alpha) in the following ways:
  
  a. 1-alpha: evenly to all of the nodes in the network.
  
  b. alpha: evenly to all the nodes they are connected to.

An individual node’s PageRank is the amount of status that it has when this process stabilizes.

To compute Pageranks, you should know about 3 concepts and how they are generated using networkx:
1. Directed Graph - It consists of a set of nodes, with a set of directed edges; each directed edge is a link from one node to another, with the direction being important.* {Syntax: g2 = nx.DiGraph()}*
2. In-degree- It is the number of links coming into a node *{Syntax:G.in_degree('A')}*
3. Out-degree- It is the number of links coming out a node *{Syntax:G.out_degree('A')}*

### Example

```python
import networkx as nx
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from IPython.display import Image
Image(filename="Graph.jpg", width = 400)
```

### PageRank Function in NetworkX
Two Parameters:
1. Graph
2. Alpha - damping parameter

```python
g3 = nx.DiGraph()
g3_edges = [("A","B"),("A","D"),("B","C"),("C","A"),("D","C")]
g3.add_edges_from(g3_edges)
nx.draw_networkx(g3, with_labels=True)
plt.show()
nx.pagerank(g3, alpha = 0.85)
```

### Finding the in-degree and out-degree

```python
print('In-degree for Node A:', g3.in_degree('A'))
print('Out-degree for Node A:', g3.out_degree('A'))
```
